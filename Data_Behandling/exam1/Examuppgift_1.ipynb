```python
# Filter the main DataFrame by 'origin' and create sub-DataFrames
df_org_1 = df[df['origin'] == 1]
df_org_2 = df[df['origin'] == 2]
df_org_3 = df[df['origin'] == 3]

# Display the first few rows of each sub-DataFrame to verify
print("Origin 1 Cars:")
print(df_org_1.head())

print("\nOrigin 2 Cars:")
print(df_org_2.head())

print("\nOrigin 3 Cars:")
print(df_org_3.head())
```

```python
# A.
def find_best_saving_origin(df):
    # Group by 'origin' and calculate the mean 'mpg'
    mean_mpg_by_origin = df.groupby('origin')['mpg'].mean()

    # Find the origin with the highest average mpg
    best_origin = mean_mpg_by_origin.idxmax()
    best_origin_mpg = mean_mpg_by_origin.max()

    return best_origin, best_origin_mpg

# Assuming 'df' is your main DataFrame
best_origin, best_origin_mpg = find_best_saving_origin(df)
print(f"The origin with the best fuel efficiency is {best_origin} with an average mpg of {best_origin_mpg:.2f}")
```

```python
# B.
def calculate_quartiles(df):
    # Calculate the quartiles
    quartiles = df['acceleration'].quantile([0, 0.25, 0.5, 0.75, 1]).to_list()
    return quartiles

# Calculate quartiles for each origin DataFrame
quartiles_org_1 = calculate_quartiles(df_org_1)
quartiles_org_2 = calculate_quartiles(df_org_2)
quartiles_org_3 = calculate_quartiles(df_org_3)

# Create a new DataFrame to display the quartile values
origin_quartiles_values = pd.DataFrame([quartiles_org_1, quartiles_org_2, quartiles_org_3],
                                       index=['origin_1', 'origin_2', 'origin_3'],
                                       columns=['min', '25%', '50%', '75%', 'max'])

# Display the new DataFrame
print(origin_quartiles_values)
```

```python
# C. 
def find_longest_history_origin(df):
    # Group by 'origin' and find the earliest model year for each origin
    earliest_year_by_origin = df.groupby('origin')['model_year'].min()

    # Find the origin with the earliest model year
    longest_history_origin = earliest_year_by_origin.idxmin()
    earliest_year = earliest_year_by_origin.min()

    return longest_history_origin, earliest_year

# Assuming 'df' is your main DataFrame
longest_history_origin, earliest_year = find_longest_history_origin(df)
print(f"The origin with the longest history in the car industry is {longest_history_origin}, starting from the year 19{earliest_year}")
```
