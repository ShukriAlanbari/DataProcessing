```python
# 1-  Read in csv file called RetailSales_BeerWineLiquor.csv in sales dataFrame
import numpy as np
import pandas as pd
df = pd.read_csv('RetailSales_BeerWineLiquor.csv')
df.head()
```

```python
# 2-  get the date in index 0 
first_date = df["DATE"][0]
first_date
```

```python
# 3-  check type of the output of #2 
type(first_date)
```

```python
# 4-  convert column 'Date' to pd.Timestamp
df['DATE'] = df['DATE'].apply(pd.Timestamp)
df
```

```python
# 5-  now check type of date into index 0
sec_date = df["DATE"][0]
type(sec_date)
```

```python
# 6-  check pandas documentation for read_csv with parse_dates argument, and read the same csv again into datafram called 
# sales_retail
sales_retail = pd.read_csv('RetailSales_BeerWineLiquor.csv',parse_dates=['DATE'])
sales_retail
```

```python
third_date = sales_retail["DATE"][0]
third_date
```

```python
# 7- check data type for index 0 in sales_retail
type(third_date)
```

```python
# 8- check documentation for resample()
```

```python
# 9- set date in sales_retail as index
sales_retail.set_index('DATE',inplace=True)
sales_retail
```

```python
# 10 - use resample() to list sales average (mean()) for each year
yearly_avg_sales = sales_retail.resample('Y').mean()
yearly_avg_sales.head()
```
