```python
import pandas as pd

# Step 1: Create df_one, df_two, df_three respectively from these dictionaries
data_1 = {'A': ['A0', 'A1', 'A2', 'A3'], 'B': ['B0', 'B1', 'B2', 'B3']}
data_2 = {'C': ['C0', 'C1', 'C2', 'C3'], 'D': ['D0', 'D1', 'D2', 'D3']}
data_3 = {'E': ['E0', 'E1', 'E2', 'E3'], 'F': ['F0', 'F1', 'F2', 'F3']}

df_one = pd.DataFrame(data_1)
df_two = pd.DataFrame(data_2)
df_three = pd.DataFrame(data_3)

# Step 2: Concat data_1 with data_2 horizontally
concat_1_2_horizontal = pd.concat([df_one, df_two], axis=1)

# Handle NaN values and reset index
concat_1_2_horizontal = concat_1_2_horizontal.dropna(axis=1, thresh=0.9*len(concat_1_2_horizontal))
concat_1_2_horizontal.reset_index(drop=True, inplace=True)

# Display the result
print("Concatenation of data_1 and data_2 horizontally:")
print(concat_1_2_horizontal)

# Step 3: Concat data_1 with data_2 vertically
concat_1_2_vertical = pd.concat([df_one, df_two])

# Handle NaN values and reset index
concat_1_2_vertical = concat_1_2_vertical.dropna(axis=1, thresh=0.9*len(concat_1_2_vertical))
concat_1_2_vertical.reset_index(drop=True, inplace=True)

# Display the result
print("\nConcatenation of data_1 and data_2 vertically:")
print(concat_1_2_vertical)

# Step 4: Concat data_1 with data_2 with data_3 vertically
concat_1_2_3_vertical = pd.concat([df_one, df_two, df_three])

# Handle NaN values and reset index
concat_1_2_3_vertical = concat_1_2_3_vertical.dropna(axis=1, thresh=0.9*len(concat_1_2_3_vertical))
concat_1_2_3_vertical.reset_index(drop=True, inplace=True)

# Display the result
print("\nConcatenation of data_1, data_2, and data_3 vertically:")
print(concat_1_2_3_vertical)

# Step 5: Concat data_1 with data_2 with data_3 horizontally
concat_1_2_3_horizontal = pd.concat([df_one, df_two, df_three], axis=1)

# Handle NaN values and reset index
concat_1_2_3_horizontal = concat_1_2_3_horizontal.dropna(axis=1, thresh=0.9*len(concat_1_2_3_horizontal))
concat_1_2_3_horizontal.reset_index(drop=True, inplace=True)

# Display the result
print("\nConcatenation of data_1, data_2, and data_3 horizontally:")
print(concat_1_2_3_horizontal)
```
