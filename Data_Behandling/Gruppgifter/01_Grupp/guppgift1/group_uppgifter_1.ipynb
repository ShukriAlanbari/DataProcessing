## Part 4
* A:    Find the books written by Pete Souza
* B:    Find the books that are under 10 dollars Ratings
* C:    Find the books that have a price between 50 and 60 dollars
* D:    Find all the books written by Kristin Hannah, Andy Weir, or Delia Owens
* E:    Find the Non Fiction books that are rated 4.9
* F:    Find the fiction book with the lowest User Rating
* G:    Find 2012's top 5 Fiction books with the most Reviews

```python
#A, Find books written by Pete Souza:
books_by_pete_souza = bestsellers_df[bestsellers_df["Author"] == "Pete Souza"]
books_by_pete_souza
```

```python
#B, Find the books that are under 10 dollars:
books_under_10_dollar = bestsellers_df[bestsellers_df["Price"] < 10]
books_under_10_dollar
```

```python
#C, Find the books between 50 and 60 dollars: 
fifty_to_sixty =\
bestsellers_df[(bestsellers_df['Price'] >= 50) & (bestsellers_df['Price'] <= 60)]

fifty_to_sixty
```

```python
#D, Find all books written by "Kristin Hannah","Andy Weir","Delia Owens":
authors_3 = ["Kristin Hannah","Andy Weir","Delia Owens"]

kristin_andy_delia = bestsellers_df[bestsellers_df['Author'].isin(authors_3)]
kristin_andy_delia
```

```python
#E, Find non-fiction books and rating over 4.9:
non_fic_and_4_9 = bestsellers_df[(bestsellers_df["Genre"] == "Non Fiction")&\
    (bestsellers_df["User Rating"] == 4.9)]
non_fic_and_4_9
```

```python
#F, find fiction book with lowest user rating:
fic_book = bestsellers_df[bestsellers_df["Genre"] == "Fiction"]
lowest_rating_book = fic_book.nsmallest(1, 'User Rating')
lowest_rating_book
```

```python
#G, Find 2012 top 5 fiction books with the highest User Rating:
books_2012 = bestsellers_df[(bestsellers_df['Year'] == 2012)&\
                           (bestsellers_df["Genre"] == "Fiction")]

top_2012 = books_2012.sort_values(by=['User Rating', 'Reviews'],
                                   ascending=[False,False])

top_5 = top_2012.head(5)
top_5
```
